```
#Example from https://www.youtube.com/watch?v=gOXoFDrseis
#import libraries
import pandas as pd
import numpy as np
from sklearn import linear_model

from sklearn.model_selection import train_test_split
```

```
#Load Boston Housing dataset from sklearn/datasets and print it
from sklearn.datasets import load_boston
boston = load_boston() # it is an object

print(boston)
```

```
# transform the dataset into a data frame
# data= independent variables, x - variables
# features names(dependent variables)
# target = the price of the houses, y - variable

df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
 
```

```
# Print some statistics 
df_x.describe()
```

```
# initializa the linear regression model

reg = linear_model.LinearRegression() 
# Split the dataset into 67% train and 33% testing data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.03, random_state =42)
```

```
# Train our linear regression model
reg.fit(x_train, y_train)
```

```
# Print the coefficients/weights for each feature/column of our model
print(reg.coef_) # f(x1, x2,... ,xd) = m1*x1 +...+ md*xd + b, where m1,m2, ...,md, b are coeffs
```

```
y_pred=reg.predict(x_test)
print(y_pred) # Looks like matrix form Ax=y, where x,y are two vectors
```

```
# print the actual values
print(y_test)
```

```
#print relative errors
print(y_pred-y_test/y_test)
```

```
# Print the Mean Square Error (MSE)
print(np.mean((y_pred-y_test)**2))
```

```
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
